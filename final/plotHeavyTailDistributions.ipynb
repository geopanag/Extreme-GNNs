```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
```

```python
pubmed = pd.read_csv("./pubmed_degree.txt", header=None)
cora   = pd.read_csv("./cora_degree.txt", header=None)
amazon = pd.read_csv("./amazon_degree.txt", header=None)
```

```python
plt.figure(figsize=(10,8))
plt.hist(pubmed[0], bins=100, density=False )
plt.yscale('log')
plt.xscale('linear')
plt.yticks(size=30)
plt.xticks(size=30, rotation=90)
plt.xlabel("degree", size=40)
plt.tight_layout()
plt.savefig("pubmed_degree_distribution.pdf", format="pdf")
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(cora[0], bins=100, density=False )
plt.yscale('log')
plt.xscale('linear')
plt.yticks(size=30)
plt.xticks(size=30, rotation=90)
plt.xlabel("degree", size=40)
plt.tight_layout()
plt.savefig("cora_degree_distribution.pdf", format="pdf", bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.hist(amazon[0], bins=100, density=False )
plt.yticks(size=30)
plt.xticks(size=30, rotation=90)
plt.xlabel("degree", size=40)
plt.yscale('log')
plt.xscale('linear')
plt.tight_layout()
plt.savefig("amazon_degree_distribution.pdf", format="pdf")
plt.show()
```
